## Resultados elección presidencial Chile 2021
Habiendo limpiado los datos del SERVEL, ofrecemos aquí un par de visualizaciones de los datos.

```python
import pandas as pd
from IPython.display import HTML

v1 = pd.read_csv('DATOS/presidencial_2021/vuelta1.csv'); x1=v1.groupby('candidato').agg({'votos':'sum'}).reset_index().sort_values('votos', ascending=False)
v2 = pd.read_csv('DATOS/presidencial_2021/vuelta2.csv');x2=v2.groupby('candidato').agg({'votos':'sum'}).reset_index()

HTML('<TABLE><TR><TD>Primera Vuelta</TD><TD>Segunda Vuelta</TD></TR><TR><TD>%s</TD><TD>%s</TD></TR></TABLE>' %(x1.to_html(), x2.to_html()))
```

```python
cruce = pd.read_csv('DATOS/presidencial_2021/cruce_V1_V2.csv')
cruce
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(15,5))
sns.heatmap(cruce.corr().tail(2).round(2), cmap='RdYlGn', annot=True, annot_kws={'weight':'bold', 'size':18});
plt.title('Correlaciones entre votos de primera y segunda vuelta', fontsize=18);
```
